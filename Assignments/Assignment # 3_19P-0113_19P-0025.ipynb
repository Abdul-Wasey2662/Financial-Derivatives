```python
#importing necessary libraries that will be required throughout the analysis.

import pandas_datareader.data as web
import datetime as dt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.dates as mdates
from matplotlib import style
pd.options.mode.chained_assignment = None
```

```python
#setting the tickers(the stocks we are analysing) and gathering the data for 5 years.

start = dt.datetime(2016, 1, 1)
end = dt.datetime(2021, 12, 31)
tickers = ['DIS', 'SONY', 'WBD']
```

```python
#reading data for our 3 stocks(Disney , Sony , WarnerBrosDiscovery) from finance.yahoo and storing it.

data = web.DataReader('DIS', 'yahoo', start, end) 
data['DIS'] = data['Adj Close']

disney = data[['DIS']]
disney['SMA20'] = disney['DIS'].rolling(20).mean()
disney['SMA120'] = disney['DIS'].rolling(120).mean()
disney.head(5)

##############################################################

data = web.DataReader('SONY', 'yahoo', start, end) 
data['SONY'] = data['Adj Close']

sony = data[['SONY']]
sony['SMA20'] = sony['SONY'].rolling(20).mean()
sony['SMA120'] = sony['SONY'].rolling(120).mean()
sony.head(5)

##############################################################

data = web.DataReader('WBD', 'yahoo', start, end) 
data['WBD'] = data['Adj Close']

warnerbros = data[['WBD']]
warnerbros['SMA20'] = warnerbros['WBD'].rolling(20).mean()
warnerbros['SMA120'] = warnerbros['WBD'].rolling(120).mean()
warnerbros.head(5)

##############################################################

display(disney)
display(sony)
display(warnerbros)
```

```python
#Plotting Disney.

style.use('seaborn-dark')
plt.figure(figsize=(12, 8))
ax1 = plt.subplot2grid((7, 1), (0, 0), rowspan = 4, colspan = 1, title = 'Disney')
ax2 = plt.subplot2grid((7, 1), (5, 0), rowspan = 2, colspan = 1, sharex = ax1)

ax1.plot(disney['DIS'], label = 'Price', alpha = 0.75)
ax1.plot(disney['SMA20'], label = 'SMA20')
ax1.plot(disney['SMA120'], label = 'SMA120')

ax2.bar(data.index, data['Volume'], label = 'Volume')

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y, %m'))

ax2.set_xlabel('Date (Year - Month)')

ax1.legend()
ax2.legend(loc = 'upper right')
```

```python
#Plotting Sony.

plt.figure(figsize=(12, 8))
ax1 = plt.subplot2grid((7, 1), (0, 0), rowspan = 4, colspan = 1, title = 'Sony')
ax2 = plt.subplot2grid((7, 1), (5, 0), rowspan = 2, colspan = 1, sharex = ax1)

ax1.plot(sony['SONY'], label = 'Price', alpha = 0.75)
ax1.plot(sony['SMA20'], label = 'SMA20')
ax1.plot(sony['SMA120'], label = 'SMA120')

ax2.bar(data.index, data['Volume'], label = 'Volume')

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y, %m'))

ax2.set_xlabel('Date (Year - Month)')

ax1.legend()
ax2.legend(loc = 'upper right')
```

```python
#Plotting Warner Bros. Discovery

plt.figure(figsize=(12, 8))
ax1 = plt.subplot2grid((7, 1), (0, 0), rowspan = 4, colspan = 1, title = 'Warner Bros. Discovery')
ax2 = plt.subplot2grid((7, 1), (5, 0), rowspan = 2, colspan = 1, sharex = ax1)

ax1.plot(warnerbros['WBD'], label = 'Price', alpha = 0.75)
ax1.plot(warnerbros['SMA20'], label = 'SMA20')
ax1.plot(warnerbros['SMA120'], label = 'SMA120')

ax2.bar(data.index, data['Volume'], label = 'Volume')

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y, %m'))

ax2.set_xlabel('Date (Year - Month)')

ax1.legend()
ax2.legend(loc = 'upper right')
```

```python
disney['price_yesterday'] = disney['DIS'].shift(1)
sony['price_yesterday'] = sony['SONY'].shift(1)
warnerbros['price_yesterday'] = warnerbros['WBD'].shift(1)

disney['Change'] = disney['DIS'] / disney['price_yesterday']
sony['Change'] = sony['SONY'] / sony['price_yesterday']
warnerbros['Change'] = warnerbros['WBD'] / warnerbros['price_yesterday']

display(disney) , display(sony) , display(warnerbros)
```

```python
#Simple Moving Average.

disney['invested_SMA'] = np.where(disney['SMA20'] > disney['SMA120'], 1, 0)
sony['invested_SMA'] = np.where(sony['SMA20'] > sony['SMA120'], 1, 0)
warnerbros['invested_SMA'] = np.where(warnerbros['SMA20'] > warnerbros['SMA120'], 1, 0)


display(disney.head(5)) , display(sony.head(5)) , display(warnerbros.head(5))
```

```python
#Buy and Hold strategy.

disney['buy_and_hold'] = np.cumprod(disney['Change'])
sony['buy_and_hold'] = np.cumprod(sony['Change'])
warnerbros['buy_and_hold'] = np.cumprod(warnerbros['Change'])


display(disney.head(5)) , display(sony.head(5)) , display(warnerbros.head(5))
```

```python
sma_disney = disney[disney['invested_SMA'] == 1]
sma_sony = sony[sony['invested_SMA'] == 1]
sma_warnerbros = warnerbros[warnerbros['invested_SMA'] == 1]

sma_disney['simple_moving'] = np.cumprod(sma_disney['Change'])
sma_sony['simple_moving'] = np.cumprod(sma_sony['Change'])
sma_warnerbros['simple_moving'] = np.cumprod(sma_warnerbros['Change'])

display(sma_disney.head(5)) , display(sma_sony.head(5)) , display(sma_warnerbros.head(5))
```

```python
sma_disney = sma_disney.drop(columns = ['buy_and_hold'])
sma_sony = sma_sony.drop(columns = ['buy_and_hold'])
sma_warnerbros = sma_warnerbros.drop(columns = ['buy_and_hold'])
```

```python
disney['EWM12'] = disney['DIS'].ewm(span =12, adjust = False).mean()
disney['EWM26'] = disney['DIS'].ewm(span =26, adjust = False).mean()

sony['EWM12'] = sony['SONY'].ewm(span =12, adjust = False).mean()
sony['EWM26'] = sony['SONY'].ewm(span =26, adjust = False).mean()

warnerbros['EWM12'] = warnerbros['WBD'].ewm(span =12, adjust = False).mean()
warnerbros['EWM26'] = warnerbros['WBD'].ewm(span =26, adjust = False).mean()

display(disney.head(5)) , display(sony.head(5)) , display(warnerbros.head(5))
```

```python
disney['invested_EWM'] = np.where(disney['EWM12'] > disney['EWM26'], 1, 0)
sony['invested_EWM'] = np.where(sony['EWM12'] > sony['EWM26'], 1, 0)
warnerbros['invested_EWM'] = np.where(warnerbros['EWM12'] > warnerbros['EWM26'], 1, 0)

display(disney.head(5)) , display(sony.head(5)) , display(warnerbros.head(5))
```

```python
ewm_disney = disney[disney['invested_EWM'] == 1]
ewm_sony = sony[sony['invested_EWM'] == 1]
ewm_warnerbros = warnerbros[warnerbros['invested_EWM'] == 1]

display(ewm_disney.head(5)) , display(ewm_sony.head(5)) , display(ewm_warnerbros.head(5))
```

```python
#Exponential Weighted Moving Average.

ewm_disney['expo_weighted_moving'] = np.cumprod(ewm_disney['Change'])
ewm_sony['expo_weighted_moving'] = np.cumprod(ewm_sony['Change'])
ewm_warnerbros['expo_weighted_moving'] = np.cumprod(ewm_warnerbros['Change'])

display(ewm_disney.head(5)) , display(ewm_sony.head(5)) , display(ewm_warnerbros.head(5))
```

```python
#Comparison(in terms of graph) of different strategies for all the stocks we were working with.

plt.figure(figsize=(14, 10))

ax1 = plt.subplot2grid((10, 1), (0, 0), rowspan = 3, colspan = 1, title = 'Comparison')
ax2 = plt.subplot2grid((10, 1), (4, 0), rowspan = 3, colspan = 1, sharex = ax1)
ax3 = plt.subplot2grid((10, 1), (8, 0), rowspan = 3, colspan = 1, sharex = ax1)


ax1.plot(disney['buy_and_hold'], label = 'Buy and Hold (Disney)')
ax1.plot(sma_disney['simple_moving'], label = 'Simple Moving Average (Disney)')
ax1.plot(ewm_disney['expo_weighted_moving'], label = 'Exponential Weighted Moving Average (Disney)')


ax2.plot(sony['buy_and_hold'], label = 'Buy and Hold (Sony)')
ax2.plot(sma_sony['simple_moving'], label = 'Simple Moving Average (Sony)')
ax2.plot(ewm_sony['expo_weighted_moving'], label = 'Exponential Weighted Moving Average (Sony)')


ax3.plot(warnerbros['buy_and_hold'], label = 'Buy and Hold (Warner Bros. Discovery)')
ax3.plot(sma_warnerbros['simple_moving'], label = 'Simple Moving Average (Warner Bros. Discovery)')
ax3.plot(ewm_warnerbros['expo_weighted_moving'], label = 'Exponential Weighted Moving Average (Warner Bros. Discovery)')

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y, %m'))

ax1.legend(fontsize = 8)
ax2.legend(fontsize = 8)
ax3.legend(fontsize = 8)
plt.show()
```
